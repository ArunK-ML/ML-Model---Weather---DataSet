```python
import pandas as pd

df = pd.read_csv("/content/Weather.csv")
df
```

```python
df.info()
```

```python
pd.set_option("display.max_columns", None)
```

```python
df["pickup_datetime"] = pd.to_datetime(df["pickup_datetime"], format = "%Y-%m-%d %H:%M:%S")
```

```python
df['date'] = df['pickup_datetime'].dt.date
df['month'] = df['pickup_datetime'].dt.month
df['time'] = df['pickup_datetime'].dt.time
```

```python
df['date']
```

```python
df['hour'] = df['pickup_datetime'].dt.hour
df['minute'] = df['pickup_datetime'].dt.minute
```

```python
df.info()
```

```python
df['day'] = df['pickup_datetime'].dt.day
```

```python
df.rename(columns={
    'tempm': 'Temp (C)',
    'tempi': 'Temp (F)',
    'dewptm': 'Dew Point (C)',
    'dewpti': 'Dew Point (F)',
    'hum': 'Humidity (%)',
    'wspdm': 'Wind Speed (km/h)',
    'wspdi': 'Wind Speed (mph)',
    'wgustm': 'Wind Gust (km/h)',
    'wgusti': 'Wind Gust (mph)',
    'wdird': 'Wind Dir (Degrees)',
    'wdire': 'Wind Dir (Compass)',
    'vism': 'Visibility (km)',
    'visi': 'Visibility (miles)',
    'pressurem': 'Pressure (mb)',
    'pressurei': 'Pressure (in)',
    'windchillm': 'Wind Chill (C)',
    'windchilli': 'Wind Chill (F)',
    'heatindexm': 'Heat Index (C)',
    'heatindexi': 'Heat Index (F)',
    'precipm': 'Precipitation (mm)',
    'precipi': 'Precipitation (in)',
    'conds': 'Weather Condition',
    'icon': 'Weather Icon',
    'fog': 'Fog Present (1/0)',
    'rain': 'Rain Present (1/0)',
    'snow': 'Snow Present (1/0)',
    'hail': 'Hail Present (1/0)',
    'thunder': 'Thunder Present (1/0)',
    'tornado': 'Tornado Present (1/0)',
    'date': 'Date',
    'month': 'Month',
    'time': 'Time',
    'hour': 'Hour',
    'minute': 'Minute'
}, inplace=True)
```
